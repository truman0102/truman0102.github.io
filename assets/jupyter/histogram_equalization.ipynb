```python
import numpy as np
import matplotlib.pyplot as plt
L = 256
```

```python
r = range(L)
p = np.random.random(L)
p = p / np.sum(p)
plt.bar(r, p)
plt.xlabel('Input intensity')
# for i, v in enumerate(p):
#     plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()
```

```python
cdf = np.cumsum(p)
convert = np.round(cdf * (L-1)).astype(int)
ps = np.zeros(L)
for i in range(L):
    ps[convert[i]] += p[i]
plt.bar(r, ps)
plt.xlabel('Output intensity')
# for i, v in enumerate(ps):
#     plt.text(i, v, f"{v:.2f}", ha='center', va='bottom')
plt.show()
```
